```python
car_list={}
def money(time): #요금 계산 함수
    base, add, max_money=3000,1000,30000
    lest_hour=time//60
    lest_time=time%60
    money=base+add*((lest_time//10)+(lest_hour-1)*6)
    if lest_hour<=1:
        money=base
    elif money>=max_money:
        money=max_money
    print(f"주차 시간은{lest_hour}시간 {lest_time}분 입니다.주차 요금은 {money}입니다.")
def in_car(): #입차 함수
    while True:
        car_num=input("입차할 차량번호를 입력해주세요. ex. 173하6456 >>>")
        if car_num in car_list:
            print("\n이미 주차된 차량입니다.다시 입력해주세요.\n")
        else:
            break
    while True:
        time=input("입차 시간을 입력해주세요. ex. 14:20 >>>").split(":")
        hour,minute=int(time[0]),int(time[1])
        sum_minute=hour*60+minute
        if hour >20 or hour < 9 or minute > 60 or minute < 0:
            print("\n잘못된 입력입니다.다시 입력해주세요.\n")
        else:
            break  
    car_list[car_num]=sum_minute
def out_car(): #출차 함수
    while True:
        car_num=input("출차할 차량번호를 입력해주세요. ex. 173하6456 >>>")
        if car_num not in car_list:
            print("\n잘못된 입력입니다.다시 입력해주세요.\n")
        else:
            break
    while True:
        time=input("출차 시간을 입력해주세요. ex. 14:20 >>>").split(":")
        hour,minute=int(time[0]),int(time[1])
        sum_minute=hour*60+minute
        if hour >20 or hour < 9 or minute > 60 or minute < 0 or sum_minute< car_list[car_num]:
            print("\n잘못된 입력입니다.다시 입력해주세요.\n")
        else:
            break
    a=sum_minute-(car_list[car_num])
    del car_list[car_num]
    return a


from prettytable import PrettyTable                                    # 시트를 출력되는 모튤을 설치하여 갖고 왔습니다.         
table = PrettyTable(["기본(1시간)","초과시 10분기준", "일일최대요금"]) 
table.add_row([3000,1000,30000])                                    
print(table)
print("-"*80)
print("-"*80)
print("\nShare 공유 주차장 찾아주셔서 감사합니다.저희 주차장 영업 시간은 9시부터 20시까지 입니다.\n") 
print("-"*80)
print("-"*80)

while True:
    select=input("\n----------메뉴----------\n1.\t입차\n2.\t출차 및 요금 계산\n3.\t주차현황\n사용하실기능을 선택해주십시요 ex. 1/2/3 \n----------------------\n")
    if select=="1":
        in_car()
    elif select=="2":
        money(out_car())
    elif select=="3":
        if car_list=={}:
            print("\n주차된 차량 없습니다.")
        else:
            for i in car_list.keys():
                print(i)
    else:
        print("\n잘못된 입력입니다.다시 입력해주세요.\n")
```
